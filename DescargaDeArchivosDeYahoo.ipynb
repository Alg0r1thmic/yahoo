```python
import datetime as dt
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web 
import csv
import math
style.use('ggplot')
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

from ipywidgets import IntProgress
from IPython.display import display
import time
```

```python
def getTickersName(df):#retorna el nombre de las columnas
    tickersNameList=[]
    for row in df.iterrows():
        index,data = row
        tickersNameList.append(data.tolist())
        break
    tickersNameList=tickersNameList[0]
    #for i in range(len(tickersNameList)-1):
     #   if(str(tickersNameList[i])=='nan'):
      #      tickersNameList.pop(i)
        
    
    return tickersNameList
def customDeleteNan(j,df):#elimina los nan de la columna j y retorna la lista limpia de los tickers de esa columna
    tem=[]
    listaFinal=[]
    for row in df.iterrows():
        index,data = row
        tem.append(data.tolist())
    for i in range(len(tem)):
        if(str(tem[i][j])!='nan'):
            if(("D:\\USA\\" in tem[i][j])==False):
                listaFinal.append(tem[i][j])
    return listaFinal
def deleteAllNan(df):#elimina los nan de todas las columnas y devuelve la lista completa de estos
    tem=[]
    listaFinal=[]
    for row in df.iterrows():
        index,data = row
        tem.append(data.tolist())
    for i in range(len(tem)):
        for j in range(len(tem[0])):
            if(str(tem[i][j])!='nan'):
                if(("D:\\USA\\" in tem[i][j])==False):
                    listaFinal.append(tem[i][j])
    return listaFinal
def saveDataSpecificTickers(tickers,df):
    currentYear=datetime.now().year
    currentMonth=datetime.now().month
    currentDay=datetime.now().day
    start=dt.datetime(2006,12,29)
    end =dt.datetime(currentYear,currentMonth,currentDay)
    max_count=len(tickers)
    f = IntProgress(min=0, max=max_count) # instantiate the bar
    display(f) # display the bar
    for i in range(len(tickers)):
        df = web.DataReader(tickers[i],'yahoo',start,end)
        f.value += 1 # signal to increment the progress bar
        time.sleep(.1)
        saveDir='D:/USA/DescargasYahoo/'+tickers[i]+'.csv'
        df.to_csv(saveDir)
def saveDataFromAllTickers(df):
    tickers=deleteAllNan(df)
    currentYear=datetime.now().year
    currentMonth=datetime.now().month
    currentDay=datetime.now().day
    start=dt.datetime(2006,12,29)
    end =dt.datetime(currentYear,currentMonth,currentDay)
    max_count=len(tickers)
    f = IntProgress(min=0, max=max_count) # instantiate the bar
    display(f) # display the bar
    for i in range(len(tickers)):
        df = web.DataReader(tickers[i],'yahoo',start,end)
        f.value += 1 # signal to increment the progress bar
        time.sleep(.1)
        saveDir='D:/USA/DescargasYahoo/'+tickers[i]+'.csv'
        df.to_csv(saveDir)
def listToDic(L):
    Ln=['none','SAVE ALL TICKERS']
    L=Ln+L
    #lis={'none':-1,'All Tickers':-1}
    lis={L[i]:i for i in range(len(L))}
    return lis
def loadCsvAndDeleteFirstColRow():
    df=pd.read_excel(r"D:/USA/Maestro.xlsx")
    df.drop(df.columns[[0]], axis=1, inplace=True)#elimina la primera columna de numeros
    df.drop(df.index[:1], inplace=True)#Elimina la primera fila de Nombres
    return df
def loadCsvAndGetColumName():
    df=pd.read_excel(r"D:/USA/Maestro.xlsx")
    df.drop(df.columns[[0]], axis=1, inplace=True)#elimina la primera columna de numeros
    tickerName=df.columns.values.tolist()
    #print(l)
    #tickerName=getTickersName(df) #se obtiene los nombres de las columnas
    df.drop(df.index[:1], inplace=True)#Elimina la primera fila de Nombres
    tickerNameList=listToDic(tickerName)
    return tickerNameList
def saveData(x):
    df=loadCsvAndDeleteFirstColRow()
    if(x==0):
        display("NONE")
    if(x==1):
        saveDataFromAllTickers(df)
        lis=deleteAllNan(df)
        display("TERMINO EXITOSAMENTE")
        print(lis)
    if(x>1):
        lis=customDeleteNan(x-2,df)
        saveDataSpecificTickers(lis,df)
        display("TERMINO EXITOSAMENTE")
        print(lis)
```

```python
def main():
    tickerNameList=loadCsvAndGetColumName()
    interact(saveData,x=tickerNameList)
if __name__ =="__main__":
    main()
```
